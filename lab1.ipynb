```python
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
# load the famous iris data
irisRaw = load_iris()
```

```python
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

```python
print(iris)
```

# Ex. 7

```python
min_value = iris["petal length (cm)"].min()
print("Max value: ",min_value)
print("Indexes: ",iris[iris["petal length (cm)"]==min_value].index.values)
```

```python
max_value = iris["petal length (cm)"].max()
print("Max value: ",max_value)
print("Indexes: ",iris[iris["petal length (cm)"]==max_value].index.values)
```

# Ex. 8

```python
print("    STANDARD DEVIATION:\nsepal length (cm): ",iris["sepal length (cm)"].std(),
     "\nsepal width (cm): ", iris["sepal width (cm)"].std(),
     "\npetal length (cm): ", iris["petal length (cm)"].std(),
     "\npetal width (cm): ", iris["petal width (cm)"].std(),
     "\ntarget: ", iris["target"].std())
```

# Ex. 9

```python
average_value = iris["sepal length (cm)"].mean()
print(average_value)
```

```python
print("Rows for which the length of the flower cup (sepal length) is greater than the average length: ",iris[iris["sepal length (cm)"]>average_value].index.values)
```

# Histograms

```python
column = iris["sepal length (cm)"]
plt.hist(column,bins = 50, facecolor = 'g')
plt.xlabel('length (cm)')
plt.ylabel('Number of cases')
plt.title('Sepal length (cm)')
plt.show()
```

```python
column = iris["sepal width (cm)"]
plt.hist(column,bins = 50, facecolor = 'g')
plt.xlabel('width (cm)')
plt.ylabel('Number of cases')
plt.title('Sepal width (cm)')
plt.show()
```

```python
column = iris["petal length (cm)"]
plt.hist(column,bins = 50, facecolor = 'g')
plt.xlabel('Length (cm)')
plt.ylabel('Number of cases')
plt.title('Petal length (cm)')
plt.show()
```

```python
column = iris["petal width (cm)"]
plt.hist(column,bins = 50, facecolor = 'g')
plt.xlabel('width (cm)')
plt.ylabel('Number of cases')
plt.title('Petal width (cm)')
plt.show()
```
